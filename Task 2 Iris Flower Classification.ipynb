### Label Encoding For Species Variable

```python
df['species']=df['species'].astype('category')
df['species']=df['species'].cat.codes
```

#### Split Data into x and y variable

```python
x=df.drop(['species'],axis=1)
y=df[['species']]
```

### Split Data into train and test 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

### KNN Classifier Model

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
# Predict
y_pred_train_knn=knn.predict(x_train)
y_pred_test_knn=knn.predict(x_test)
```

```python
## Evaluate
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,precision_score

# Predict
accuracy_knn_test=accuracy_score(y_test,y_pred_test_knn)
accuracy_knn_train=accuracy_score(y_train,y_pred_train_knn)

print("KNN Classifier - Train Accuray : " ,accuracy_knn_train)
print('\n','--------------------------------------------','\n')

print('KNN Classifier - Train Accuracy : ', accuracy_knn_test)
```

```python
# Cross Validation
from sklearn.model_selection import cross_val_score
train_accuracy_knn=cross_val_score(knn,x_train,y_train,cv=10)
cross_val_train_knn=train_accuracy_knn.mean()
test_accuracy_knn=cross_val_score(knn,x_test,y_test,cv=10)
cross_val_test_knn=test_accuracy_knn.mean()

print('KNN Classifier after Cross Validation Train Accuracy :',cross_val_train_knn)
print('**********************' * 4)
print('KNN Classifier after Cross Validation Test Accuracy:',cross_val_test_knn)
```

```python
accuracy=[]
for i in range(1,10):
    knn=KNeighborsClassifier(n_neighbors=i)
    score=cross_val_score(knn,x,y,cv=10)
    accuracy.append(score.mean())
    
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,10),accuracy,color='blue',linestyle='dashed',marker='*',markerfacecolor='red',markersize=10)
plt.title('Accuracy Vs K-Points')
plt.xlabel('K-Points')
plt.ylabel('Accuracy')
plt.show()
```

- From plot,it is noticed that highest accuracy is at n=9.
